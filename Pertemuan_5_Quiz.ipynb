```python
print("-Soal 5-")
print("Menghitung Jumlah Sapi Per 7 hari dan Berapa hari jika sapi lebih dari 500")

total_sapi_7hari = 0
sapi_hari_ini = 3

for hari in range(1, 8):
    total_sapi_7hari += sapi_hari_ini
    sapi_hari_ini += 2

print("Jumlah sapi dalam 7 hari pertama =", total_sapi_7hari)

total_sapi = 0
sapi_hari_ini = 3
hari = 0

while total_sapi < 500:
    total_sapi += sapi_hari_ini
    sapi_hari_ini += 2
    hari += 1

print("Jumlah hari yang dibutuhkan untuk mencapai ≥ 500 sapi =", hari)
print("Total sapi setelah", hari, "hari =", total_sapi)
```
