```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sales_abiyyu_df = pd.read_csv("/content/sales.csv")
```

```python
# Ringkasan statistik
sales_abiyyu_df.describe()

# Rata Rata Untuk kolom price per unit
rata_rata_price_per_unit = sales_abiyyu_df["price_per_unit"].mean()
print("Rata Rata Price Per Unit: ", rata_rata_price_per_unit)

# Rata Rata Untuk Kolom Quantity
rata_rata_quantity = sales_abiyyu_df["quantity"].mean()
print("Rata Rata Quantity: ", rata_rata_quantity)

# Rata Rata Untuk kolom Total Price
rata_rata_total_price = sales_abiyyu_df["total_price"].mean()
print("Rata Rata Total Price: ", rata_rata_total_price)

# Modus untuk kolom product id
modus_product_id = sales_abiyyu_df["product_id"].mode()
print("Modus Product Yang Sering Di Pesan: ", modus_product_id)
```

Disini untuk bagian Praktik 1 yang dimana menampilkan detail detail statistik secara deskriptif untuk dari data tabel sales.csv, disini terdapat pada ringkasan statistik yang dimana isinya ada rata-rata kolom price per unit, rata-rata untuk kolom quantity, ada rata-rata untuk kolom total price dan juga ada modus untuk product yang paling sering terjual, jadi disini bisa dilihat dari hasil table penjualan produk yang sering terjual adalah product id nomor 78, dan rata rata harga per unitnnya 103,5, rata rata quantity barang yang terjual itu 1,9 di mana ini masih di bilang sedikit untuk penjualan.

```python
# Histogram Distribusi Harga Per Unit

sales_abiyyu_df["price_per_unit"].hist()
plt.title("Distribusi Harga Per Unit")
plt.xlabel("Harga Per Unit")
plt.ylabel("Frekuensi")
plt.show()
```

Ini untuk melihat Diagram distribusi harga per unit jadi untuk harga per unit disini bisa terlihat belum cukup merata, Disini terlihat untuk harga per unit ada yang frekuensinya sampai ke 700 ada juga yang cuman 100, jadi disini terlihat untuk distribusi penyebaran harga per unitnya

```python
# Histogram Distribusi Quantity

sales_abiyyu_df["quantity"].hist()
plt.title("Distribusi Quantity")
plt.xlabel("Quantity")
plt.ylabel("Frekuensi")
plt.show()
```

Untuk yang di bagian Ini distribusi quantity yang distribusi quantity itu adalahseberapa banyak barang dari diagram ini terlihat hanya tiga sebaran yaitu diantara 1, 2 dan 3.

```python
# Histogram Distribusi Total Price

sales_abiyyu_df["total_price"].hist()
plt.title("Distribusi Total Price")
plt.xlabel("Total Price")
plt.ylabel("Frekuensi")
plt.show()
```

Di bagian ini adalah distribusi untuk total price yang dimana total fresh ini merujuk kepada harga akhir yang dibeli jadi perkalian antara quantity dan price per unit dan dapatlah total price, sini menunjukkan bahwa total price yang di bagian kanan sudah sedikit merata anda moon di bagian yang kiri terlalu tinggi, Yang artinya penyebaran untuk total harganya itu tidak merata



```python
# Korelasi & Kovarian price per unit dengan quantity

# Korelasi
corr_matrix = sales_abiyyu_df[["price_per_unit", "quantity"]].corr()
print("Korelasi: ", corr_matrix)

# Kovarian
cov_matrix = sales_abiyyu_df[["price_per_unit", "quantity"]].cov()
print("Kovarian: ", cov_matrix)
```

Hubungan korelasi antara dua parameter price_per_unit dan quantity yang dimana hasilnya korelasinya positif tapi sangat kecil, yang berarti hubungan nya tidak saling mempengaruhi, dan tidak saling berhubungan, untuk hubungannya linier walaupun harga per unit naik, belum tentu quantitynya ikut naik juga

```python
# Korelasi & Kovarian price per unit dengan total price

# Korelasi
corr_matrix = sales_abiyyu_df[["price_per_unit", "total_price"]].corr()
print("Korelasi: ", corr_matrix)

# Kovarian
cov_matrix = sales_abiyyu_df[["price_per_unit", "total_price"]].cov()
print("Kovarian: ", cov_matrix)
```

Ini di buat juga sebagai perbandingan dari data sebelumnya, hubungan antara dua parameter ini price_per_unit dan total price, yang hasilnya positif namun cukup kecil, ini saling berhubungan namun tidak terlalu kuat dimana ini linear hubungannya semakin jika ada perubahan di harga per unit, totalnya juga berubah tapi sesuai dengan kenaikan harga per unitnya, total harga ini akan berubah drastis tergantung qty yang terjual

```python
# Korelasi & Kovarian price quantity dengan total price

# Korelasi
corr_matrix = sales_abiyyu_df[["quantity", "total_price"]].corr()
print("Korelasi: ", corr_matrix)

# Kovarian
cov_matrix = sales_abiyyu_df[["quantity", "total_price"]].cov()
print("Kovarian: ", cov_matrix)
```

Ini adalah menampilkan korelasi dan kovarian antara quantity dan total price yang dimana, korelasi antara kedua parameter ini saling mempengaruhi satu sama lain di angka 0,9 hampir menyentuh 1, untuk kovariannya bernilai positif semakin meningkat quantity yang terjualmaka semakin meningkat juga total harga.

```python
# Kolom kategori quantity
sales_abiyyu_df["Kategori_Quantity"] = sales_abiyyu_df["quantity"].apply(lambda x: "Banyak" if x >= 3 else "Sedikit")
sales_abiyyu_df


# Rata-rata populasi per kategori
group_mean = sales_abiyyu_df.groupby("Kategori_Quantity")["total_price"].mean()
print(group_mean)


# Agregasi lengkap
group_agg = sales_abiyyu_df.groupby("Kategori_Quantity").agg({
    "total_price": ["mean", "min", "max"],
    "quantity": ["mean", "min", "max"]
})
group_agg
```

Ini menampilkan kategori dari sebuah data yang dibuat untuk melihat dan mengkategorikan banyak atau sedikit dari data yang telah kita analisis.
